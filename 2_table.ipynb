```python
from datetime import date
```

```python
import pandas as pd
import numpy as np
customer=pd.DataFrame({'cust_id':[100,101,102,103,104],'cust_nm':['tim','bill','ravi','pallavi','prithvi'],'cust_city':['Chennai','Bangalore','Chennai','Bangalore','Mumbai'],'cust_state':['TN','KA','TN','KA','MH'],
                 'cust_dob':['22-02-2000','04-01-2000',np.nan,'10-04-1998','18-10-1997']})
inquiry=pd.DataFrame({'inquiry_id':[1,2,3,4,5,6,7],'inquiry_date':['25-08-21','25-08-21','14-08-21','15-08-21','20-08-21','21-08-21','22-08-21'],'inquiry_type':['Internet','Voice','Internet','Internet','Voice','Internet','Voice'],'Inquiry_type':['New Connection','Upgrade','Data Upgrade','Line Down message','New Connection','Line Down message','Data Upgrade'],'emp_id':[3455,3455,3400,3400,3455,3400,2400],'cust_id':[np.nan,101,102,102,np.nan,103,102]})
```

```python
customer['cust_dob']=pd.to_datetime(customer['cust_dob'])
customer
```

```python
inquiry
```

```python
inquiry['inquiry_date']=pd.to_datetime(inquiry['inquiry_date'])
inquiry
```

```python
#Find all the customers who raised an inqury in the month of August 2021
customer[inquiry['inquiry_date'].dt.month==8].merge(inquiry,left_on='cust_id',right_on='cust_id',how='inner')
```

```python
#Find customers who never raised any inquiry
customer[['cust_nm','cust_id']][~customer['cust_id'].isin(inquiry['cust_id'])]
```

```python
#Display customername, inquiry_date, inquiry for the all customers who are from the state of KA
both=customer.merge(inquiry)
both[['cust_nm','inquiry_date','inquiry_id']][both['cust_state']=='KA']
```

```python
#Find all the inquieries which are not from our customers
inquiry[~inquiry['cust_id'].isin(customer['cust_id'])].merge(customer,left_on='cust_id',right_on='cust_id',how='left')
```

```python
#Display inquiry_type and the number of inquieries we got in the current year
inquiry[['inquiry_type']][inquiry['inquiry_date'].dt.year==date.today().year],'count=',inquiry['inquiry_type'][inquiry['inquiry_date'].dt.year==date.today().year]
```

```python
#Display Customer_name, inquiry_type, inquiry_date for the customer inqueries. Show the customers who did not do any inquiry as well in the result set
customer[~customer['cust_id'].isin(inquiry['cust_id'])].merge(inquiry,on='cust_id',how='left')
```

```python
"""Display the required data as below from the above set of tables
InquiryDoneBy       Number of Inquries
Customers           5
Non Customers       2"""
```

```python
inquiry_report=pd.DataFrame({'InquiryDoneBy':['Customers','Non Customers'],
                             'Number of Inquries':
                             [inquiry[~inquiry['cust_id'].isin(customer['cust_id'])].merge(customer,left_on='cust_id',right_on='cust_id',how='left')['inquiry_id'].count(),
                              inquiry[inquiry['cust_id'].isin(customer['cust_id'])].merge(customer,left_on='cust_id',right_on='cust_id')['inquiry_id'].count()]})
```
